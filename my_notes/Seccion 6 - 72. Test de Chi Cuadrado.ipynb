El test de la chi cuadrado se trata de un __test de la estad√≠stica muy utilizado para comparar los datos observados con los datos esperados asumiendo que los datos siguen una determinada hip√≥tesis__

En el __sentido cl√°sico del contraste de hip√≥tesis__, uno supone como __hip√≥tesis nula__ que los datos que tienen entre manos __siguen una determinada distribuci√≥n.__
- una normal, una uniforme, una binomial 
- que __los datos siguen una distribuci√≥n y en base a esa hip√≥tesis calcular cu√°les ser√≠an los valores esperados de distribuci√≥n.__

Entonces:
- __E_i__: los parametros esperados o las caracter√≠sticas esperadas
- __O_i__: los par√°metros observados o las caracter√≠sticas observadas
-
- b√°sicamente restando esos dos valores podemos calcular las discrepancias entre lo observado y lo esperado seg√∫n la teor√≠a.
    - De nuevo __se elevan al cuadrado esas discrepancias__ para que las positivas no contrarresten las negativas, sino que todas cuenten y al final cada uno de ellas se divide entre el valor esperado
    -  __el sumatorio__ se lleva a cabo en todas y cada uno de los puntos de entrada, todos los datos que quer√©is comparar.
    
__Las diferentes formas que tenemos de aplicar el test de la chi cuadrado__:
-	independencia (comparar dos varibles)
-	homogeneidad (comparar dos variables)
-	bondad de ajuste.

#### Bondad de ajuste
Para comprobar si los datos vienen de una fuente de verdad que sigue una distribuci√≥n conocida normal o no-normal.

B√°sicamente __se observa si los datos que tenemos tienen cierto desv√≠o o cierta tendencia hacia un extremo u otro comparados con los datos esperados.__

- __Si los datos no se ajustan a la distribuci√≥n que dicen seguir entonces la prueba de la bondad de ajuste falla.__ Entonces el test de chi cuadrado nos puede ayudar.

- En este caso la __hip√≥tesis nula__ ser√≠a: los datos siguen una distribuci√≥n en contra de los datos, no siguen dicha distribuci√≥n.

Para el caso de __comparar dos variables x e y__ tenemos dos pruebas diferentes uno es el test de __homogeneidad__ y el otro es el test de __independencia.__

#### Independencia y homogeneidad

Nos ayuda a demostrar una posible relaci√≥n o independencia entre dos variables, un par√°metro de entrada y una salida.

__Para llevarse a cabo__:
- Se asume de nuevo la independencia, la __hip√≥tesis nula__ es "existe independencia entre estas dos variables" y __se calculan los valores esperados__ que tendr√≠amos en caso de independencia.

- __Se calcula el par√°metro de la chi cuadrado__ exactamente igual:
    - __restando las observaciones menos las esperadas al cuadrado y dividiendo por las esperadas__ 
- __En caso de rechazar la hip√≥tesis nula__ se sugiere que existe una relaci√≥n entre dos variables.

Esta relaci√≥n no es por azar, no es casualidad, sino que realmente __hay herramientas estad√≠sticas o hay razones de peso de la estad√≠stica para garantizar que realmente son diferentes__. No son independientes de que exista una dependencia.


Tambi√©n se utiliza __para comprobar si los datos son demasiado buenos para ser verdad (si no han sido trucados).__ Normalmente en un experimento aleatorio uno no espera que los valores sean caprichosos, que vayan tendencia dos hacia la hip√≥tesis que uno quiere comprar.

#### Grados de libertad

El chi cuadrado tiene lo que llamamos __grados de libertad__ que b√°sicamente modela una forma diferente.

La imagen superior muestra un par de cuadrados dependiendo del grado de libertad desde 1 hasta 10:
-	para igual a 1 esta tendencia va hacia la izquierda y a medida que van incrementando los grados de libertad se va aplanando.

__Realmente se necesita saber exactamente el n√∫mero de grados de libertad__. 
- En el caso de una __bondad de ajuste__: siempre ser√° una menos un grado de libertad menos que las categor√≠as que tiene la variable aleatoria que est√°is intentando modelar
- En el caso de un __contraste de independencia o de homogeneidad__: ser√° el n√∫mero de filas menos 1 por el n√∫mero de columnas menos 1.
