La __correlación__ es una idea estadística muy importante y básica sale cuando necesitamos encontrar una relación entre dos variables.


- el concepto de correlación es una __premisa básica para crear un modelo de datos, para poder llevar a cabo un análisis predictivo__, 
    - la correlación es la capacidad de encontrar el factor, __la función que permite predecir la salida del modelo.__
    
- Una buena correlación entre dos variables sugiere que existe algún tipo de dependencia entre ellas si cambia una, el cambio se verá reflejado también en la otra. 
- La correlación no es más que __una relación matemática, lo que llamamos una función, que se establece entre dos variables debido a su propia naturaleza de relación matemática__, ya sea una operación aritmética exponencial o demás. 
- Nos permite predecir los valores de salida,  lo cual es súper útil para los __modelos predictivos.__

Esta __relación matemática puede ser de cualquier tipo__, en particular si X e Y son dos variables que están relacionadas, uno siempre podrá escribir Y como función de X ó X como función de Y. (Típicamente siempre ponemos en función de x por notación estándar.)

De éstas __hay dos que son muy particulares__,
- __si ‘f’ es una función lineal entonces ‘a’ y ‘b’ que son coeficientes de un polinomio de grado 1__, es decir X e Y están correlacionados de forma lineal. 
    - Podría tener distribuciones con grados cuadráticos formando una parábola, grados cúbicos y demás, pero la más típica es la correlación lineal entre variables.

- La __correlación exponencial o logística__, que básicamente es cuando aparecen una de las variables en el exponente de una base. 
    - Típicamente os he puesto aquí el número ‘e’, pero no tendría por qué ser exactamente el número ‘e’, podría ser 2 elevado a algo, o 5 elevado a algo, 10 elevado a algo, o la operación contraria, que es la logarítmica que es cuando sale un logaritmo.
    
- Tanto uno como otro también aparecen muchísimo en relaciones entre variables y dan lugar a algunos de los fundamentos básicos de la creación de los modelos predictivos.

Sin embargo, __¿cómo podemos categorizar el grado de correlación?__ 
- No puede decir simplemente el grado es bueno o es malo, __hay que cuantificarlo.__

Para ello existe el __coeficiente de correlación de Pierson__, que se calcula con esta fórmula que tenéis aquí. 
- Simplemente es cada uno de los valores X e Y,  se restan respectivamente a la media, la media de X y de Y por separado,  y se dividen por la raíz cuadrada de sumar sólo las X por un lado, sumar sólo las Y por otro y multiplicar dichos resultados.

- Es una fórmula que se calcula automáticamente, __la importancia es aprender a identificar qué significa ese valor__. 
    - __El valor de la correlación siempre estará en el rango entre menos 1 y 1__, fijaros que ahí tenéis abajo el dibujito menos uno y medio tienes el cero y a la derecha tenéis el 1.

#### Correlación positiva
Una __correlación positiva__ significa que:
- el coeficiente de correlación de Pierson estará __entre 0 y 1__ 
- habrá una __relación directa__ entre esas dos variables directas, que si una variable incrementa la otra también lo hará.
- __si ese coeficiente es muy cercano a cero__, incluso en la literatura encontraréis que si es más pequeño que __0’6 o 0’64__ es una correlación que es __mera causalidad y que por tanto no hay que darle demasiada importancia.__
- __A partir de 0’6 , 0’7,  0’8 y ya tendiendo hacia 1 es cuando esa correlación cobra mucho sentido.__

#### Correlación negativa
Por otro lado, una __correlación negativa__ será aquella que:

- su coeficiente de correlación irá __de -1 a 0__ 
- significará que tiene una __relación inversa__ entre las dos variables. Por tanto, cuando una incremente la otra descenderá y si una vez decrece la otra incrementará. 
- __Para valores inferiores en este caso a  -0’6, -0’7__ y además es cuando la correlación negativa __cobra algo de sentido para valores pequeños o cercanos a cero__, apenas le damos importancia 

__Cuanto mayor es el valor de la correlación mayor es la relación que se establece entre las dos variables que se están realizando.__

__Una correlación fuerte parece sugerir que existe algún tipo de relación que se puede utilizar para crear un modelo predictivo.__

__Si tenemos una variable que podemos medir y otra que podemos inferir en base a una correlación significa que nos iría muy bien para crear un modelo.__

__Que exista correlación no implica necesariamente que haya un único factor que explique la causalidad.__ 
- Podría haber varios, podría haber diferentes e incluso podría ser que la correlación no fuera de uno de estos modelos explicados en lo que fuere de algún otro tipo de modelo que se nos escapa que fuera muy difícil de elaborar matemáticamente, entonces hay una frase muy típica es que la correlación no necesariamente implica la causalidad. __"Que dos variables estén correlacionadas no significa que una cause a la otra o viceversa, sino que simplemente van de la manita."__

```python
import numpy as np
```

```python
# Coeficiente de Correlacion de Pearson

### Numerador de la ecuación: Para TV y ventas (sales)
data_ads["corrn"] = ((data_ads["TV"] - np.mean(data_ads["TV"]))
                    * (data_ads["Sales"] - np.mean(data_ads["Sales"])))
```

```python
data_ads.head()
```

```python
# Calculos para el denominador

# Para la TV
data_ads["corr1"] = (data_ads["TV"] - np.mean(data_ads["TV"]))**2

# Para las ventas(Sales)
data_ads["corr2"] = (data_ads["Sales"] - np.mean(data_ads["Sales"]))**2

data_ads.head()
```

```python
# Sumatorios y division entre numerador y denominador
corr_pearson = sum(data_ads["corrn"]) / np.sqrt(sum(data_ads["corr1"])
                                               * sum(data_ads["corr2"]))
print(corr_pearson)
```

Da un valor positivo y encima creciente, esto significa que cuanto más se gasta en TV, más se incrementan las ventas

```python
def correlation_coeficient(df, var1, var2):
    # Numerador (sin sumatorio)
    df["corrn"] = ((df[var1] - np.mean(df[var1]))
                   * (df[var2] - np.mean(df[var2])))
    
    # Denominador (sin sumatorio)
    # var1
    df["corr1"] = (df[var1] - np.mean(df[var1]))**2
    # var2
    df["corr2"] = (df[var2] - np.mean(df[var2]))**2
    
    # Sumatorios y division entre numerador y denominador
    corr_pearson = sum(df["corrn"]) / np.sqrt(sum(df["corr1"])
                                               * sum(df["corr2"]))
    
    
    #print(corr_pearson)
    return corr_pearson
```

```python
correlation_coeficient(data_ads,"TV", "Sales")
```
