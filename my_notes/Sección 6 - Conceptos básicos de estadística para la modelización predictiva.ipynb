Asumimos precisamente una hipótesis generalmente relacionada con el valor que tiene que tomar un estimador. Si es exactamente igual a un número o si es más grande o más pequeño que un cierto valor.

Típicamente esta suposición que hace el estadístico lo llamamos __hipótesis nula__ y lo que __intentamos es verificar si esa afirmación es o no es cierta aplicando una serie de reglas que tienen que ver con la distribución normal.__

- A la hipótesis contraria la que no está contemplada en la hipótesis nula, el caso contrario, se llama la __hipótesis alternativa.__

__¿Cuál es cierto la hipótesis nula o la hipótesis alternativa?__

+ __Hipotesis Nula__
    +  Es nuestra premisa inicial y lo que __queremos defender que es cierto hasta que se demuestre lo contrario.__ : H Sub-Zero.

    + La vamos a defender a capa y espada. Queremos que sea cierta 

+ __Hipotesis Alternativa__
    + Es algo que no estamos muy seguros, lo proponemos como alternativa. 
    + Lo único que hacemos es contradecir en muchos casos la hipótesis nula y nos la vamos a quedar si no queda más remedio.


Cuando alguien está haciendo __búsqueda cuantitativa para calibrar por ejemplo el valor de un estimador__, que __conocemos el valor del parámetro__ porque ya lo teníamos calculado anteriormente o hemos hecho los cálculos teóricos, ese valor, precisamente el conocimiento del que hace la investigación, es el valor que se coloca o que se toma como hipótesis nula.
+ El valor históricamente ha valido esto.
+ Mientras que __el nuevo valor o la nueva hipótesis__ que uno quiere tomar que quiere probar si es cierto o es falso se toma como hipótesis alternativa.

__Ejemplo:  Edad promedio de los habitantes de Barcelona.__
-	Pues bueno nos vamos a los __índices históricos__ de la demografía en Barcelona y uno se encuentra con que la edad promedio en Barcelona es inferior a 38 años. Esto puede servirnos como __hipótesis nula__
-	Mientras que en su __propia investigación__ haciendo las muestras que habíamos tomado pues justamente le sale un __valor típicamente mayor.__
-	Si la edad de 38 años o al menos es como la afirmación era menor de 38 años no hay ningún problema. __El problema está cuando haces su investigación en lugar de salir menos de 38 le sale 42 y medio__
    - __En términos estadísticos estamos diciendo una mentira.__
    - Hay suficientes evidencias como para tomar que la hipótesis alternativa es la buena que la edad no es menos de 38 sino que la edad promedio en Barcelona ha aumentado.
    
+ __Esto es de lo que trata precisamente el contraste de hipótesis en afirmarlo de forma estadística más allá de decir sí o no de acuerdo.__

### El teorema central del límite (TCL)
Vamos a tener que recurrir a __el teorema central del límite__ y a utilizar dos estadísticos:  
+ la zeta la función normal
+ la t de Student.

De modo que a la hora de validar un contraste de hipótesis u otro tendremos que utilizar un estadístico de contraste u otro dependiendo de una serie de características.

#### Hipótesis Nula
Independientemente del tipo de contraste de hipótesis que queramos hacer, __asumimos que el valor que aparece en la hipótesis nula es Mu Sub-Zero.__

+ Bien sea que queremos validar que es exactamente igual ese valor o mayor o menor nos ponemos en entredicho dicha afirmación.
    + Tomamos una muestra aleatoria simple de tamaño 100 o de tamaño 1000 de un tamaño fijo N de ocurrencias del suceso y calculamos el promedio del parámetro, como podría ser el promedio de edad, el tiempo promedio que tardan en llevar una pizza a casa, los ingresos promedios por ejemplo de una familia o de una ciudad o de la gente que trabaja en la universidad
        - __Ejemplo:__ Justeat dice que en menos de 20 minutos lo tienes en casa, pues llamamos 100 veces al Justeat y comprobamos a ver si es verdad.
        
    + Calculamos el promedio de lo que queramos le llamamos Mu Sub-X y según el teorema central del límite, la distribución de las medias de la población del parámetro estudiado, tiende a una distribución que va a ser normal.
    
#### El estadístico Zeta
Se calcula para __convertirlo a una distribución normal__, una distribución que tenga media cero y desviación típica uno,  de modo que la nuestra tendrá una distribución normal Mu y Sigma
-	En principio mu y sigmas desconocidos, nosotros habremos estimado a través de una muestra aleatoria simple de tamaño n un estadístico Mu-sub X y un estadístico Sigma-sub X, que en principio tendrán que ver con la de la población y básicamente las normalizo.

Si restamos promedio a la medida estudiada y dividimos por la desviación típica de la muestra esto se convierte en una normal 01.
-	Esto tiene un problemita, pero si yo quiero calcular esto pues debo tomar como desviación típica, la desviación típica de la propia población que en principio no la conozco.

Hemos dicho que la medida de la población podemos tomar como media la que estoy poniendo en entresijo, es decir,  si es cierto o no es cierto que la población tiene medíamos Sub-Zero, pero la desviación típica fijaros que necesito Sigma para poder hacer Sigma divido por raíz de n,  lo cual es un problema porque dependiendo de los casos podría ser que yo no tuviera Sigma, es decir: 
-	mu lo tenemos porque es el Sub-Zero que estoy poniendo en entredicho,  
-	n lo tengo porque es el tamaño de la muestra,  
-	puedo tener X barra porque es el promedio de la muestra que he tomado.
-	con Sigma existen dos casos:
    + cuando conocemos  la desviación típica (caso 1)
    + cuando no conocemos la desviación típica (caso 2)
    
    
#### Caso 1: 
+ __Ejemplo:__ experiencia de un investigador, estudios previos, registros, etc.
+ Nuestra hipótesis nula es conocida Mu Sub-Zero es conocida porque la estoy poniendo en entredicho, N es conocida porque yo he tomado la muestra, el promedio de la misma también la puedo estimar porque para eso he tomado la muestra aleatoria simple y la desviación estándar es conocida en este caso por la experiencia del investigador.

+ En este caso es el caso más específico y de hecho el único en que se puede aplicar un Z test, lo que llamamos un test de normalidad. Porque básicamente lo tenemos todo para afirmar que la distribución sigue una distribución normal y que el valor normalizado del valor operado como lo tenéis en la primera de las dos casillas corresponde con un estadístico Z o lo que llamamos un Z valor, un valor de la normal

#### Caso 2:
+ Tal vez es la primera vez que nuestro investigador o nuestro estadístico lleva a cabo dicho análisis, no hay datos históricos o es la primera vez que a alguien se le ocurre estudiar esto o incluso el número de personas o el número de sucesos estudiados es tan pequeño que ni siquiera se puede asumir normalidad.

+ Entonces en este caso la estimación de la media se puede llevar a cabo como antes pero la desviación típica tiene que movilizarse porque la desviación típica es conocida.
+ __Ejemplo:__ el más típico es las notas de los estudiantes en un examen, la edad de una población o cosas que pueden ir cambiando y que no es seguro que nosotros tengamos movilizadas
+ En este caso tanto el promedio como la desviación estándar., los dos se convierten en parámetros desconocidos y por tanto la expresión a evaluar. en el caso del tema central del límite pues tiene dos incógnitas una es la Mu y la otra es Sigma.
    + Ambas son desconocidas de nuevo se utiliza la que hemos estimado en la que ponemos en entredicho la media, pero necesitamos estimar Sigma.
+ Esto significa que el tema central del límite deja de ser válido porque la distribución que surge en estos casos es una distribución que se llama la __t de Student o el parámetro o el estadístico t__.
+ El valor estándar en este caso se llama T valor en lugar de Z valor y el test que se lleva a cabo es un test en lugar de un Z test.

La  distribución de la desviación estándar debe estimarse a la vez que estimamos el promedio y en el caso de que el número de muestras sea suficientemente grande pues podemos garantizar dicha convergencia a la T de Student.

Llamemos al estadístico que aproxima a la desviación típica S, entonces ese se calcula como el sumatorio de cada uno de los valores menos el promedio de la muestra, es decir cómo se desvía cada valor con respecto de la media de la muestra se hace al cuadrado porque si lo sumara simplemente daría cero porque cada valor sumado con las diferencias de la media unos se pasan por arriba otros se pasan por abajo entonces se acabarían anulando las diferencias, para ello se elevan al cuadrado y se divide por en menos 1.
-	Es una variante de la definición clásica de desviación típica.

Entonces bajo estas condiciones, si la media de la muestra le resto de la población y divido por este parámetro estimado divido raíz de n, sigue precisamente una distribución, que no es normal sino que es una t de student.

Las diferencias entre los dos casos, es el tipo de distribución que sigue el primero directamente y de forma rotundamente clásica si una distribución normal y se calcula utilizando un Z valor o Z test.
El segundo actúa con una distribución de t de Student y se calcula un t-valor asociado al análisis que queremos llevar a cabo.

Fijaros que estos estadísticos se convierten en el estadístico Zeta y el estadístico t, que son los parámetros que nos van a ayudar en nuestro contraste de hipótesis.

Estamos en el supuesto en que he podido calcular dos estadísticos de contraste, o bien el Z o bien el T, y básicamente en función de lo que me ha dado dicho valor,  podré afirmar si la hipótesis que quería poner en entredicho es cierta o es falsa.

__Habíamos dejado en entredicho la hipótesis nula__ y en base a las tomas que habíamos hecho __habíamos calculado un estadístico de contraste__ que bien podía ser: 
-	de tipo normal un Z-test
-	de tipo t de Student, un T-test

__La función de distribución o de probabilidad acumulada__ que básicamente consistía en ir contando la probabilidad que existía desde un punto hacia atrás en la distribución.

+ El área debajo de una curva de cualquier distribución, ya sea una uniforma normal o una t de Student, siempre el máximo que puede sumar es 1.

Entonces si echamos un vistazo a la __imagen superior__ fijaros que en este caso tenemos el dibujo de una distribución normal pero nada quita que pudiera ser una t de Student.

Supongamos que:
-	__tenemos z1 y z2:__ dos estadísticos Z correspondientes a dos valores de la variable aleatoria en cuestión podrían ser dos estadísticos que nos han salido al hacer una normalización con el teorema central del límite 
-	y  __P1 y P2__ las probabilidades encerradas por debajo de la curva justo a la izquierda.

Fijaros que:
-	la probabilidad de que la distribución X tome un valor menor que Z1 es exactamente igual a P1
    - P1 es la probabilidad de que la variable aleatoria toma un valor menor o igual a z1
-	la probabilidad de que la variable aleatoria X tome un valor menor a Z2, la hemos llamado P2, es el área encerrada debajo de la curva exactamente a la izquierda de estos valores
    - P2 es la probabilidad de que la variable aleatoria tome un valor menor igual que Z2 respectivamente.
-	aquí __la variable aleatoria la llamó X__ ya que si es una normal la llamaremos Z pero como X puede ser tanto normal como t de Student.
-	la suma de las probabilidades tiene que dar uno, __el área total bajo la curva es uno:__
    - si lo que queda por debajo de Z1 es P, __lo que queda por encima de Z1 es 1 menos P1 y lo que queda por encima de Z2 es 1 menos P2.__
    - esto va a ser muy __importante cuando tengamos que calcular intervalos de confianza y de llevar a cabo el contraste de hipótesis__
    
Para distribuciones bien definidas como es la distribución normal evidentemente uno puede __definir un intervalo dentro del cual el valor de la probabilidad de la variable aleatoria va a caer.__
-	eso es lo que se llama un __intervalo de confianza__


#### El intervalo de confianza

```python
Image(filename="/Users/snopv/PycharmProjects/DataScience_Python/my_notes/images/71_2.png",  width=600, height=300)
```

En el intervalo, la longitud de ese intervalo, lo grande que es,  se le llama el nivel de confianza.
-	Por __ejemplo__: para una distribución normal que tenga media Mu y desviación típica Sigma 
    + el valor de la variable aleatoria, por ejemplo, en un intervalo mu -3 sigma mu +3 Sigma implica irme tres desviaciones típicas a la izquierda y tres a la derecha del valor central 
        - tiene un 99 por ciento de probabilidad, por tanto __es un intervalo de confianza al 99 por ciento de nivel de confianza__.
-	__El nivel de confianza__:
    - Es la probabilidad con la que nosotros estamos seguros de que la variable aleatoria va a caer ahí dentro en cualquier experimento aleatorio__
    -	Para cualquier estimador, esencialmente cualquier variable aleatoria que siga una distribución normal se puede definir un intervalo de confianza si nosotros fijamos a priori el nivel de confianza, la probabilidad, cómo de grande queremos que sea dicho intervalo.
    -	Uno puede pensar en el nivel de confianza o en los intervalos de confianza como __umbrales o límites del valor de aceptación que nosotros estamos dispuestos a aceptar para defender la hipótesis nula como verdad verdadera.__

Si yo hago en torno al valor que estoy defendiendo de la hipótesis nula, recordad aquel Sub-Zero, un intervalo de confianza centrado en él y el valor de mi estimador cae dentro del rango, estadísticamente será correcto mantener la hipótesis nula como válida.

Si por contra el estimador o bien DZ o bien T que hemos llevado a cabo se sale de ese umbral de confianza entonces no habrá más remedio que aceptar la hipótesis alternativa.

__Para definir el intervalo de confianza__ hay que decidir cuál es el nivel de confianza o probabilidad que estamos dispuestos a aceptar. 
-	esta probabilidad normalmente se suele definir por el estadístico que está llevando a cabo el análisis a priori y antes de hacer el test.

#### Nivel de significación 
Si nosotros llamamos a la probabilidad p entonces normalmente en lugar de fijarlo en p se suele definir como __uno menos p__, entonces:
-	uno menos P no es el nivel de confianza, sino que define lo que llamamos el __nivel de significación  y típicamente lo representamos con la letra alfa.__
    - representa la probabilidad de que la hipótesis nula no sea cierta.
-	Entonces si la definimos nosotros mismos para cada uno de los test que llevemos a cabo suele ser del orden de 0,1; 0,05; 0,01; 0,005; son números pequeños y es la probabilidad de que la hipótesis nula sea falsa 
    - fijando unos niveles unas probabilidades tan bajas, uno __cuando ya se encuentra que el valor no cae dentro del intervalo de confianza que había fabricado pues la hipótesis nula no se sustenta por ningún lado.__
        - De ahí que sea __más útil definir el nivel de significación en lugar del nivel de confianza__, pero simplemente uno es el contrario el opuesto del otro.

#### p-valor
También un concepto muy importante que tenemos que tener en cuenta es el __p-valor o el valor probabilístico del estadístico de contraste__. 
- si el estadístico de contraste que nos ha devuelto o Z-test o un T-test, el Z valor del valor es el que tenéis marcado abajo como z cero,  __se define el valor como la probabilidad de que la función de distribución supere dicho valor.__
    -	__el p-valor es la probabilidad máxima de caernos fuera de caer por encima de dicho valor.__
    
Ahora que __tenemos tanto z-valor como el p-valor__ obtenidos, siempre asumiendo que la hipótesis nula es verdadera, __la única forma que hay de que la hipótesis nula sea aceptada es que el Z- valor esté fuera de la región de rechazo__, es decir, que la hipótesis nula será verdadera única y exclusivamente __si el p-valor es superior al nivel de confianza__ (y esto en el dibujo se ve fantásticamente bien).
-	Este sería un __contraste de hipótesis lateral por la derecha__ 
    -	Por tanto, si el Z valor cae fuera de la región de rechazo o lo que es lo mismo el __p-valor es más grande que el nivel de significación, entonces aceptaremos la hipótesis nula.__
    -	Si no pues no nos quedará más remedio que rechazarla y tener que aceptar como buena la hipótesis alternativa.

#### Tipos de test
Debido a la asimetría y a la naturaleza de las distribuciones normales y demás existen tres __tipos posibles de test__:
-	por la izquierda o de cola por la izquierda.
-	por la derecha o de cola por la derecha 
-	o el análisis de dos colas.

La imagen muestra los dos primeros, el de la cola por la izquierda y el de la derecha que normalmente.

#### Cola por la izquierda
Una cola por la izquierda es el caso donde __la hipótesis alternativa es menor.__
-	el test que se hace en este caso da lugar a una distribución donde la importancia está en el análisis de la cola izquierda de la distribución 
-	arriba del todo aceptaremos la hipótesis nula y rechazaremos la alternativa o bien si el valor supera a nivel de significación o bien si el estadístico de contraste es superior al estadístico que corresponde al nivel de significación pertinente

__Siempre tendremos dos formas de concluir si aceptamos o rechazamos:__
-	O bien comparando los estadísticos de contraste con el estadístico de significas del nivel de significación pertinente 
-	O bien comparando el nivel de significación con el valor.

Por el contrario, en una cola por la izquierda, rechazaremos en el caso contrario si el estadístico de contraste, fijaros en este caso es inferior al estadístico con el nivel de significación, fijaros que está por debajo el valor en este caso del nivel de significación establecido por tanto en este caso rechazaría la hipótesis nula y tendríamos que aceptar la alternativa.

#### Cola por la derecha
En el caso de una cola por la derecha el caso es absolutamente el contrario:
-	la hipótesis alternativa es del tipo mayor-que y en este caso el contraste de la hipótesis se centra en analizar el lado derecho de la distribución, de ahí que sea un análisis de la cola por la derecha.
-	en este caso aceptaríamos la hipótesis nula y rechazábamos la alternativa si el p-valor de nuevo es superior al nivel de significación.
    - Esto se traduce en que el estadístico de contraste es menor que el estadístico del nivel de significación, Zeta es menor que Zeta Alfa 
-	rechazábamos en caso contrario, si z es mayor que Z alfa o lo que es lo mismo, se traduce en que el nivel de significación es superior al p-valor.

#### Dos colas

En el caso de dos colas :
-	se junta el dibujo de arriba con el de abajo
-	la única diferencia es que el nivel de significación alfa se reparte en Alfa medios en la cola izquierda y Alfa medios a la derecha
    - no tenéis sólo una zeta Alfa, tenéis dos Zeta Alfa la superior y la inferior, típicamente una y otra son iguales pero opuestas, cambiadas de signo, si uno es uno coma noventa y seis el otro es menos uno coma noventa y seis.
-	aceptamos si nos encontramos dentro del intervalo de confianza.
-	rechazamos si superamos el umbral establecido en la región de aceptación bien por arriba o bien por abajo.
    - si cualquiera de los dos es tanto el de arriba como el de abajo me hacen rechazar la hipótesis nula entonces rechazamos en un contraste bilateral
