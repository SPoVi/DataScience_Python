# Carga de datos a través de la función open

```python
data3 = open(mainpath + "/" + "customer-churn-model/Customer Churn Model.txt","r")
```

```python
cols = data3.readline().strip().split(",")
# Número de columnas
n_cols = len(cols)
```

```python
# Encontrar el número de filas
counter = 0
main_dict = {}  #diccionario
for col in cols:
    main_dict[col] = []
```

```python
# Por cada línea en el dataset
for line in data3:
    values = line.strip().split(",") # Leer valor de la línea
    
    # Normalmente el range(1,5) tiene un abanico, de 1 al 5 (cuatro veces)
    for i in range(n_cols):
        # Crea un diccionario resultado de la inversa del texto
        # Lo que antes estaba en la posicion mxn pasa a estar nxm
        # Cambia el contenido de filas por columnas
        main_dict[cols[i]].append(values[i])
        
    counter += 1

print("El data set tiene %d filas y %d columnas" %(counter-1, n_cols))
```

##### Explicación del loop anterior
Si el texto tenia como datos "Nombre", "Edad", "Altura":
Linea 1: Daniel, 27 , 1.85
Linea 2: Pablo, 25, 1.75

Recoge la informacion que está en una línea de texto plano, distinguiendo cada parámetro, y guardandolos en un diccionario haciendo referencia a las columnas.

No es exacto, pero para hacerse una idea...:
diccionario = {columna1=Nombre:(Daniel, Pablo), columna2=Edad:(27,25), colummna3=Altura:(1.85,1.75)}

#### Ejercicio de descarga de datos con urllib3

Usar la librería urllib3 para leer los datos desde una URL externa, procesarlos y convertirlos a un data frame de python antes de guardarlos en un CSV local.


```python
#Creacion de función para descargar datos de URL y guardalos en un csv local

def downloadFromURL(url, filename, sep=",", delim="\n", encoding="utf-8",
                   mainpath = "/Users/snopv/PycharmProjects/DataScience_Python/my_datasets"):
    
    # Importar la libreria y hacer la conexion con la url
    import urllib3
    http = urllib3.PoolManager()
    r = http.request('GET', url)
    print("El estado de la respuesta es %d" %(r.status))
    response = r.data
    
    # El objeto response contiene un string binario, así que lo convertimos a un string
    # decodificandolo en UTF-8
    str_data = response.decode(encoding)
    
    # Dividimos el string en un array de filas, separandolo por saltos de linea
    lines = str_data.split(delim)
    
    # La primera línea contiene la cabecera, así que la extraemos 
    col_names = lines[0].split(sep)
    n_cols = len(col_names)
    
    # Generamos un diccionario vacío donde irá la información procesada desde
    # la URL externa
    counter = 0
    main_dict = {}
    for col in col_names:
        main_dict[col] = []
        
    # Procesamos fila a fila la información para ir rellenando el diccionario
    # con los datos como hicimos antes
    for line in lines:
        # Nos saltamos la primera línea que es la que contiene la cabecera
        # y que ya tenemos procesada
        if (counter > 0):
            # Dividimos cada string por las comas como elemento separador
            values = line.strip().split(sep)
            # Añadimos cada vlaor a su respectiva columna del diccionario
            for i in range(len(col_names)):
                main_dict[col_names[i]].append(values[i])                 
        counter += 1
                           
    print("El data set tiene %d filas y %d columnas"%(counter-1, n_cols))
    
    # Convertimos el diccionario procesado a Data Frame y comprobamos que los
    df = pd.DataFrame(main_dict)
    print(df.head())
    
    # Elegimos donde guardarlo (en la carpeta athletes es dnde tiene mas sentidos
    # por el contexto del análisis)
    fullpath = os.path.join(mainpath, filename)
    
    # Lo guardamos en CSV, JSON o en Excel según queramos
    df.to_csv(fullpath + ".csv")
    df.to_json(fullpath + ".json")
    df.to_excel(fullpath + ".xlsx")
    print("Los ficheros se han guardado correctamente en: " + fullpath)
    
    return df
                          
```
