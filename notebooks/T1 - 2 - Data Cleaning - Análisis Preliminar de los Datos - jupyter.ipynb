```python
data["sex"].head()
```

Como se puede observar, la variable sex es categórica: male / female

Se sustituye las categorias por columnas

```python
# Pandas facilita método, donde puedes indicar el prefijo a añadir para
# las categorías
dummy_sex = pd.get_dummies(data["sex"], prefix="sex")
```

```python
dummy_sex.head()
```

Eliminamos la columna con la variable categorica y la sustituimos por
las dummies correspondientes


```python
# Obtener la lista de las columnas de la cabecera
column_name = data.columns.values.tolist()
column_name
```

```python
# Eliminar la columna correspondiente a la categorica: "sex"
data6 = data
data6 = data6.drop(["sex"], axis = 1)
```

```python
# Añadimos las columnas dummies
data6 = pd.concat([data6, dummy_sex], axis = 1)
```

```python
# Mostrar resultados
data6.head()
```

También podemos definir una función que lo haga

```python
def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop(var_name, axis=1)
    df = pd.concat([df, dummy], axis = 1)
    return df
```

```python
data6 = data
# Probamos con la categoria de embarque
createDummies(data6, "embarked")
```
