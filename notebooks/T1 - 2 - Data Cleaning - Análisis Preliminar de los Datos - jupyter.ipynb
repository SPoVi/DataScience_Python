# Missing values

```python
# Valores nulos
pd.isnull(data["body"])
```

```python
# Valores no nulos
pd.notnull(data["body"])
```

```python
# Contarlos valores nulos o no-nulos
nulo = pd.isnull(data["body"]).values.ravel().sum()
no_nulo = pd.notnull(data["body"]).values.ravel().sum()
print("Hay %d nulos y %d no-nulos"% (nulo, no_nulo))
```

Los valores que faltan en un data set pueden venir por dos razones:
- Extracción de los datos
- Recolección de los datos

#### Borrado de valores que faltan

```python
# Axis = 0 -> Borrado de filas (eje vertical); con 1 borra columnas
# How = "all" -> Solo si todas las columnas contienen NaN
data.dropna(axis=0, how="all")
```

```python
data2 = data
```

```python
# How = "any" -> si la fila contiene algún NaN en alguna de sus columnas
data2.dropna(axis=0, how="any")
```

#### Cómputo de los valores faltantes

Existen varias opciones:
+ Rellenar con cero
+ Poner una palabra
+ Poner la media o la mediana
+ Métodos de __*fillna*__:
    - forwardfield
    - backwardfield

```python
data3 = data
```

```python
# Reemplazar por un cero
data3.fillna(0)   # El "fillna" no sobreescribe el dataset original
                  # No era necesario hacer copias del "data"  
```

```python
data4 = data
```

```python
# Reemplazar por una palabra
data.fillna("Desconocido")

# Sobrescribir el dataset:
data4 = data4.fillna("Desconocido")
```

```python
data5 = data
data5.head()
```

```python
data5["body"].head()
```

```python
# Sobreescribir colmunas seleccionadas 
data5["body"] = data5['body'].fillna(0)
data5["home.dest"] = data5["home.dest"].fillna("Desconocido")
data5.head(10)
```

```python
# Contar la cantidad de celdas nulas 
pd.isnull(data5["age"]).values.ravel().sum()
```

```python
# Rellenar con la media
data5["age"].fillna(data["age"].mean()) # No esta sobreescribiendo
```

```python
# Usar el método de fillna "fordwardfill"
# Coje el valor más cernano anterior (fila menor)
data5["age"].fillna(method="ffill")
```

```python
# Usar el método de fillna "backwardfill"
# Coje el valor más cercano posterior (fila mayor)
data5["age"].fillna(method="backfill")
```
