# Data Wrangling - La cirugía de los datos



El __data wrangling__, a veces denominada __data munging__, es el proceso de transformar y mapear datos de un dataset raw (en bruto) en otro formato con la intención de hacerlo más apropiado y valioso para una variedad de propósitos posteriores, como el análisis. 
+ Un __data wrangler__ es una persona que realiza estas operaciones de transformación.

Esto puede incluir:
- munging
- visualización de datos 
- agregación de datos
- entrenamiento de un modelo estadístico 
- otros usos potenciales... 

La oscilación de datos como proceso generalmente sigue un conjunto de pasos generales:
1.  Extrayendo los datos en forma cruda del origen de datos
2.  Dividiendo los datos en bruto usando algoritmos (por ejemplo, clasificación) o analizando los datos en estructuras de datos predefinidas
3.  Finalmente depositando el contenido resultante en un sistema de almacenamiento (o silo) para su uso futuro.

#### Subconjunto de filas con condiciones booleanas

```python
# Usuarios con Day Mins > 300
data1 = data[data["Day Mins"] > 300]
data1.shape
```

```python
# Usuarios de Nueva York (State = "NY")
data2 = data[data["State"]=="NY"]
data2.shape
```

Combinar varias condiciones

```python
# AND -> &
data3 = data[(data["Day Mins"] > 300) & (data["State"] == "NY")]
data3.shape
```

```python
# OR -> |
data4 = data[(data["Day Mins"] > 300) | (data["State"] == "NY")]
data4.shape
```
